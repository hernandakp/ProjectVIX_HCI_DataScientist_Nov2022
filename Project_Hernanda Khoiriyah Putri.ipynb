```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
list(df_dropna.columns)
```

```python
# Visualisasi presentase perbandingan contract type
df_contracttype = [np.count_nonzero(df_dropna[df_dropna.NAME_CONTRACT_TYPE == "Cash loans"]), np.count_nonzero(df_dropna[df_dropna.NAME_CONTRACT_TYPE == "Revolving loans"])]
label = ['Cash loans', 'Revolving loans']
plt.title('Contract Type')
plt.pie(df_contracttype, labels =label, radius=1.0, startangle=60, autopct='%.1f%%', shadow=True)
plt.show()
```

- Presentase pada kolom contract_type adalah 89,2% cash loans dan 10,8% revolving loans.

```python
# Visualisasi presentase CODE_GENDER
df_gender= [np.count_nonzero(df_dropna[df_dropna.CODE_GENDER == "F"]), np.count_nonzero(df_dropna[df_dropna.CODE_GENDER == "M"])]
label = ['F', 'M']
plt.title('Code Gender')
plt.pie(df_gender, labels =label, radius=1.0, startangle=60, autopct='%.1f%%', shadow=True)
plt.show()
```

- Presentase pada kolom code_gender adalah 49,1% perempuan (F) dan 50,9% laki-laki (M).

```python
# Visualisasi presentase perbandingan FLAG_OWN_CAR
df_FLAG_OWN_CAR = [np.count_nonzero(df_dropna[df_dropna.FLAG_OWN_CAR == "Y"]), np.count_nonzero(df_dropna[df_dropna.FLAG_OWN_CAR == "N"])]
label = ['Y', 'N']
plt.title('FLAG_OWN_CAR')
plt.pie(df_FLAG_OWN_CAR, labels =label, radius=1.0, startangle=60, autopct='%.1f%%', shadow=True)
plt.show()
```

- Presentase pada kolom FLAG_OWN_CAR adalah 100% client memiliki mobil (Y).

```python
# Visualisasi presentase perbandingan FLAG_OWN_REALTY
df_FLAG_OWN_REALTY = [np.count_nonzero(df_dropna[df_dropna.FLAG_OWN_REALTY == "Y"]), np.count_nonzero(df_dropna[df_dropna.FLAG_OWN_REALTY == "N"])]
label = ['Y', 'N']
plt.title('FLAG_OWN_REALTY')
plt.pie(df_FLAG_OWN_REALTY, labels =label, radius=1.0, startangle=60, autopct='%.1f%%', shadow=True)
plt.show()
```

- Presentase pada kolom FLAG_OWN_REALTY adalah 68,7% Y (memiliki rumah sendiri) dan 31,3% N (tidak memiliki rumah sendiri).

```python
# Melihat jumlah CNT_CHILDREN
CNT_CHILDREN = df_dropna['CNT_CHILDREN'].value_counts()
CNT_CHILDREN = pd.DataFrame(CNT_CHILDREN).reset_index()
CNT_CHILDREN
```

- Data menunjukkan bahwa jumlah client yang tidak memiliki anak adalah jumlah terbanyak. Sedangkan jumlah client paling sedikit adalah memiliki 5 anak, yaitu hanya 2 client.

```python
# Distribusi dari AMT_INCOME_TOTAL
f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df_dropna['AMT_INCOME_TOTAL'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df_dropna['AMT_INCOME_TOTAL'])
plt.show()
```

- Density dari distribusi pendapatan client (AMT_INCOME_TOTAL) berada di angka 200000an. 
- Data memiliki banyak outliers.

```python
# Distribusi dari AMT_CREDIT
f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df_dropna['AMT_CREDIT'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df_dropna['AMT_CREDIT'])
plt.show()
```

- Density dari distribusi jumlah kredit dari pinjaman (AMT_CREDIT) berada di angka 500000an. 
- Data memiliki banyak outliers.

```python
# Distribusi dari NAME_INCOME_TYPE
f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
sns.countplot(df_dropna['NAME_INCOME_TYPE'])
```

- Sumber pendapatan client yang terbanyak adalah working, disusul oleh commercial associate dan state servant.
